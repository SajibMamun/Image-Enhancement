```python
import cv2
from google.colab.patches import cv2_imshow
# Load an image in grayscale
image = cv2.imread('/content/00075aa8-d81a-4184-8541-b692b78d398a___FREC_Scab 3335.jpg', 0)

# Create a CLAHE object
clahe = cv2.createCLAHE(clipLimit=4.0, tileGridSize=(10, 10))

# Apply CLAHE to the image
clahe_image = clahe.apply(image)

# Display the original and CLAHE-enhanced images
cv2_imshow(image)
cv2_imshow(clahe_image)

```

```python

import numpy as np
# Define the sharpening kernel
sharpening_kernel = np.array([[0, -1, 0],
                              [-1, 5, -1],
                              [0, -1, 0]])

# Apply the kernel to the image
sharpened_image = cv2.filter2D(clahe_image, -1, sharpening_kernel)
# Display the original and CLAHE-enhanced images
cv2_imshow(clahe_image)
cv2_imshow(sharpened_image)
```
